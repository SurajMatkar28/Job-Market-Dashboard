```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("clean_jobs.csv")
df.head()
```

TOP 10 JOB TITLES

```python
top_jobs=(df["title"].value_counts().head(10))
```

```python
top_jobs
```

```python
plt.figure(figsize=(10,6))

plt.barh(
    top_jobs.index,
    top_jobs.values
)

plt.title("Top 10 Job Titles")

plt.xlabel("Count")

plt.tight_layout()

plt.show()
```

```python
top_locations=(df["location"].value_counts().head(10))
```

```python
plt.figure(figsize=(10,6))
plt.barh(top_locations.index,top_locations.values)
plt.title("Top Hiring Locations")
plt.xlabel("Number of Jobs")
plt.tight_layout()
plt.show()
```
