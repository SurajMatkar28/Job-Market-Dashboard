```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("jobs.csv")
df.head()
```

```python
df.info()
```

```python
keep=["title",
      "company_name",
      "location",
      "max_salary","med_salary","min_salary",
      "formatted_work_type","applies","views","remote_allowed"]

df=df[keep]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["title","location"])
```

```python
df.isnull().sum()
```

```python
df["company_name"]=(df["company_name"].fillna("Unknown"))
```

```python
df.isnull().sum()
```

```python
salary_df=df.dropna(subset=["med_salary"])
```

```python
salary_df["med_salary"].describe()
```

```python
df["views"]=(df["views"].fillna(df["views"].median()).astype(int))
```

```python
df["applies"]=(df["applies"].fillna(0))
```

```python
df["applies"]=(df["applies"].astype(int))
```

```python
df["remote_allowed"]=(df["remote_allowed"].fillna("Unknown"))
```

```python
salary_df=(df.dropna(subset=["med_salary"]))
```

```python
salary_df.shape
```

```python
df.isnull().sum()
```

```python
df.to_csv("clean_jobs.csv",index=False)
```
